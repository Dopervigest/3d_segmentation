```python
import os
import numpy as np
from tqdm.notebook import tqdm

import torch
import torch.nn.functional as F
from torchsummary import summary
from torch.utils.data import TensorDataset, DataLoader
from scripts.models import UNet
from scripts.dataset import H5Dataset
from scripts.utils import plot_loss, plot_accuracy

from scripts.trainer import Trainer, Classification_head

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

# Training on ACDC

```python
out_dir = './models/ACDC_Unet'

if os.path.exists(out_dir):
    print("path exists")
else:
    os.makedirs(out_dir)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
train_dataset = H5Dataset('./prepared_data/train/data.hdf5')
test_dataset = H5Dataset('./prepared_data/test/data.hdf5')

batch_size = 128
epochs = 100

train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

len(train_dataset), len(test_dataset)
```

```python
model = UNet(n_classes=3).float()
opt = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5) # выбираем оптимизатор и загружаем в него модель
criterion = F.binary_cross_entropy  # выбираем лосс
model.to(device).float() 
print()
```

```python
trainer = Trainer(model, opt, criterion, train_dataloader, test_dataloader, out_dir, epochs, device)
```

```python
trainer.fit()
```

```python
plot_loss(out_dir + '/logs.csv')
```

```python
plot_accuracy(out_dir + '/logs.csv')
```

```python
preds = []
labels = []

model = torch.load(f'{out_dir}/best.pt') # загружаем модель с лучшими весами
model.to('cpu')
model.eval() 
for x,y in iter(test_dataloader): 
    pred = model(x) 
    preds.append(pred.detach())

    labels.append(y)


preds = np.concatenate(preds, axis=0)
labels = np.concatenate(labels, axis=0)

preds = np.argmax(preds, axis=1)
labels = np.argmax(labels, axis=1)


report = classification_report(y_true=labels, y_pred=preds)
print(report)
```

```python
cm = confusion_matrix(y_true=labels,
                      y_pred=preds)


# рисуем её
plt.Figure()
disp = ConfusionMatrixDisplay(cm,)

disp.plot(values_format='')
```
